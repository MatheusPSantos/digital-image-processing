# Image annotation

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from IPython.display import Image
%matplotlib inline
```

```python
# lendo uma imagem
image = cv2.imread("./assets/apollo.jpg", cv2.IMREAD_COLOR)
# mostra imagem original
plt.imshow(image[:,:,::-1])
```

## desenhar linha na imagem

Usamos a função `cv2.line` para isso.

- img: imagem que iremos desenhar a linha
- pt1: primeiro ponto(x,y) do segmento
- pt2: segundo ponto(x,y) do segmento
- color: cor da linha a ser desenhada

Outros argumentos importantes para se saber:
- thickness: inteiro especificando a linha thickness. Valor default 1.
- lineType: Tipo da linha. Valor default é 8 que é padrão para 8-conectado. Usualmente cv2.LINE_AA (antialiased ou smoth line) é usado para lineType;

```python
imageLine = image.copy();
cv2.line(imageLine, (200, 200), (1000, 200), (0,244, 123), thickness=5, lineType=cv2.LINE_AA)
plt.imshow(imageLine[:,:,::-1])
```

## desenhando circulo

- img: imagem que receberá o circulo
- center: centro do circulo
- radius: raio da circunferência
- color: cor do circulo que será desenhado.

Outros parâmetros:
1 - thickness: Thickness do outline do cirulo (se positivo). Se negativo, resultará num circulo preenchido.
2 - lineType: Tipo de regra do círculo. É o mesmotipo do cv2.line.




```python
#desenhar circulo
imageCircle = image.copy()
cv2.circle(imageCircle, (1050, 950), 100, (124,124,124), thickness=5, lineType=cv2.LINE_AA)
plt.imshow(imageCircle[:,:,::-1])
```

## Desenhando retângulo
Usamos a função `cv2.rectangle`.
- img: Imagem em que vamos desenhar o retangulo
- pt1: Vertice do retangulo. Usualmente o topo esquerdo.
- pt2: Vertice do retangulo. Usualmente a base direita.
- color: Cor do retangulo.

Outros parâmetros:
1. thickness: Espessura do retangulo. Se nagativo, então será preenchido.
2. lineType: tipo da linha do retangulo. O mesmo tipo do cv2.line.


```python
# Desenha o retangulo
imageRectangle = image.copy()
cv2.rectangle(imageRectangle, (400, 150), (650, 1000), (120,100,200), thickness=-1, lineType=cv2.LINE_AA)
plt.imshow(imageRectangle[:,:,::-1])
```

## adicionando texto
Usamos a função `cv2.putText`.
- img: imagem onde adicionaremos o texto.
- text: texto string que será escrito.
- org: Bottom-left canto do texto string na imagem.
- fontFace: tipo da fonte.
- fontScale: escala da fonte.
- color: cor da fonte.

1. thickness: espessuar do texto.
2. lineType: Tipo da linha. Padrão é 8, usualmente temos cv2.LINE_AA.


```python
imageText = image.copy()
text = "Apollo 11 Saturn V launch, July 16, 1969"
fontScale = 2.4
fontFace = cv2.FONT_HERSHEY_PLAIN
fontColor = (125, 250, 250)
fontTchikness = 3

cv2.putText(imageText, text, (200, 1100), fontFace, fontScale, fontColor, fontTchikness, cv2.LINE_AA)
plt.imshow(imageText[:,:,::-1])
```
