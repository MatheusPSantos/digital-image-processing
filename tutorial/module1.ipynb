Trabalhando com cores.

Ler e mostrar uma imagem com cores

```python
img_color = cv2.imread("./assets/imagem2.png", 1)
# dimensão da image
print(img_color.shape)
# data type da imagem
print(img_color.dtype)
```

3 é o número de canais

## Display imagem

```python
plt.imshow(img_color)
```

Open CV utiliza um método diferente de armazenamento da informação do canal de cor. Para imagens RGB o openCV utiliza canal BGR, matplot lib espera um RGB mas recebe um BGR do openCV

```python
img_color_com_canal_reservado = img_color[:,:,::-1]
plt.imshow(img_color_com_canal_reservado)
```

## Spliting e Merging Canais de cores

```python
img_bgr = cv2.imread("./assets/imagem1.jpg", cv2.IMREAD_COLOR)
b,g,r = cv2.split(img_bgr)

# mostrar canais
plt.figure(figsize=[20,5])
plt.subplot(141); plt.imshow(r, cmap="gray");plt.title("Red Channel")
plt.subplot(142); plt.imshow(g, cmap="gray");plt.title("Green Channel")
plt.subplot(143); plt.imshow(b, cmap="gray");plt.title("Blue Channel")

# mergear canais individualmente
imgMergeada = cv2.merge((b,g,r))
# mostrar saída mergeada
plt.subplot(144);plt.imshow(imgMergeada[:,:,::-1]);plt.title("Merged output")
```

## Convertendo para diferentes colors spaces

Usa a função do opencv `cvtColor`.

```python
img_hsv = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2HSV)
h,s,v= cv2.split(img_hsv) 
# mostrar canais
plt.figure(figsize=[20,5])
plt.subplot(141); plt.imshow(h, cmap="gray");plt.title("H Channel")
plt.subplot(142); plt.imshow(s, cmap="gray");plt.title("S Channel")
plt.subplot(143); plt.imshow(v, cmap="gray");plt.title("V Channel")
plt.subplot(144); plt.imshow(img_bgr);plt.title("Original")
```

### Modificando canais
É possível modificar um canal

```python
h_new = h + 100
img_hsv_mergeada = cv2.merge((h_new, s,v))
img_hsv_rgb = cv2.cvtColor(img_hsv_mergeada, cv2.COLOR_HSV2RGB)
# mostrar canais
plt.figure(figsize=[20,5])
plt.subplot(141); plt.imshow(h_new, cmap="gray");plt.title("H Channel")
plt.subplot(142); plt.imshow(s, cmap="gray");plt.title("S Channel")
plt.subplot(143); plt.imshow(v, cmap="gray");plt.title("V Channel")
plt.subplot(144); plt.imshow(img_hsv_rgb);plt.title("Original")
```
