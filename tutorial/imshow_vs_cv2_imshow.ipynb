## Imshow do matplotlib vs imshow do cv2

```python
import cv2
import matplotlib.pyplot as plt

imagem_1 = cv2.imread("./assets/imagem1.jpg")
imagem_2 = cv2.imread("./assets/imagem2.png")

#usando matplotlib imshow
plt.imshow(imagem_1)
plt.title("imshow do matplotlib")
plt.show()

# usando imshow do opencv, mostrando por 8 sec
window1 = cv2.namedWindow("vindow 1")
cv2.imshow(window1, imagem_1)
cv2.waitKey(8000)
cv2.destroyWindow(window1)

# usando imshow, mostrando at uma tecla ser pressionada
window2 = cv2.namedWindow("window 2")
cv2.imshow(window2, imagem_1)
cv2.waitKey(0)
cv2.destroyWindow(window2)

window3 = cv2.namedWindow("window 3")
Alive = True
while Alive:
    # usando imshow do opencv at a tecla 'q' ser precionada
    cv2.window(window3, imagem_1)
    keypress = cv2.waitKey(1)
    if keypress == ord('q'):
        Alive = False
cv2.destroyWindow(window3)

cv2.destroyAllWindows()
stop = 1
```
